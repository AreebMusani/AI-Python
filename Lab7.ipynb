# ACTIVITY 5

Write a Python program to find the real and imaginary parts of an array of complex numbers

```python
import numpy as np
complex_num = np.array([-1 + 9j, 2 - 77j,
                        31 - 25j, 40 - 311j,
                        72 + 11j])
for i in range(len(complex_num)):
      print("{}. complex number is {}".format(i + 1, complex_num[i]))
      print("The real part is: {}".format(complex_num[i].real))
      print("The imaginary part is: {}\n".format(complex_num[i].imag))
```

# ACTIVITY 10

Write a Python program to find common values between two arrays using intersect1d ( ). Take 
array1= [ 1 30 40 60 80, 90] and array2= [50,10,30,40] as input.

# ACTIVITY 11

Write a Python program to draw individual line charts of the financial data of a company given in 
Table A-11 between October 3, 2022 to October 7, 2022 using subplot for each attribute.

Date     Open       High Low Close
10-03-22 774.25     776.065002 769.5 772.559998
10-04-22 776.030029 778.710022 772.890015 776.429993
10-05-22 779.309998 782.070007 775.650024 776.469971
10-06-22 779 780.47998 775.539978 776.859985
10-07-22 779.659973 779.659973 770.75 775.080017

```python
import matplotlib.pyplot as plt
import pandas as pd
data = {
    'Date': ['10-03-22','10-04-22','10-05-22','10-06-22', '10-07-22'],
    'Open': [774.25, 775.030029, 779.309998, 779, 779.659973],
    'High': [776.065003, 778.710022, 782.070007, 780.47998, 779.659973],
    'Low': [769.5, 772.890015, 775.650024, 775.539978, 770.75],
    'Close': [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]
}
df = pd.DataFrame(data)
df['Date'] = pd.to_datetime(df['Date'], format='%m-%d-%y')
df.set_index('Date', inplace=True)
fig,axes = plt.subplots(nrows=4, ncols=1, figsize=(4,10))
df['Open'].plot(ax=axes[0], color='red')
df['High'].plot(ax=axes[1], color='blue')
df['Low'].plot(ax=axes[2], color='green')
df['Close'].plot(ax=axes[3], color='purple')

axes[0].set_title('Open Price')
axes[1].set_title('High Price')
axes[2].set_title('Low Price')
axes[3].set_title('Close Price')
for ax in axes:
    ax.set_xlabel('Date')
    ax.set_ylabel('Price')
plt.tight_layout()
plt.show()
```
