# ACTIVITY # 1

```python
import heapq

# Define the state graph as a dictionary of dictionaries
graph = {
    'S': {'A': 2, 'B': 2},
    'A': {'G': 2},
    'B': {'G': 3},
}

# Define the heuristic function as the straight-line distance to the goal
heuristic = {
    'S': 3,
    'A': 2,
    'B': 1,
    'G': 0
}

def a_star_search(start, goal, graph, heuristic):
    pq = []  # min-priority queue
    closed = set()
    
    g = {start: 0}
    f = {start: heuristic[start]}
    path = {start: []}
    heapq.heappush(pq, (f[start], start))
    
    while pq:
        (f_curr, curr) = heapq.heappop(pq)
        if curr == goal:
            return f[curr], [start] + path[curr]
        closed.add(curr)
        for next, cost in graph[curr].items():
            if next not in closed:
                g_next = g[curr] + cost
                if next not in f or g_next + heuristic[next] < f[next]:
                    g[next] = g_next
                    f[next] = g_next + heuristic[next]
                    path[next] = path[curr] + [next]
                    heapq.heappush(pq, (f[next], next))
      
        

f, path = a_star_search('S', 'G', graph,heuristic)

# Print the results
print('Shortest Path:', path)
print('Total Cost:', f)
```

# ACTIVITY # 6

```python
import heapq
graph = {
    'S': {'A': None, 'B': None},
    'A': {'G': None},
    'B': {'C': None},
    'D': {'G': None, 'C': None},
    'G': {'D': None},
    'C': {'D': None},
}
heuristic = {
    'S': 7,
    'A': 6,
    'B': 5,
    'C': 4,
    'D': 2,
    'G': 0,
}

def best_first_search(start, goal, graph, heuristic):
    pq = [(heuristic[start], start)]
    explored = set()
    path = {start: []}
    
    while pq:
        curr_h, curr = heapq.heappop(pq)
        
        if curr == goal:
            return curr_h, path[curr] + [curr]
        
        explored.add(curr)
        
        for neighbor in graph[curr].keys():
            if neighbor not in explored:
                h = heuristic[neighbor]
                heapq.heappush(pq, (h, neighbor))
                path[neighbor] = path[curr] + [curr]
    
    return None

heuristic, path = best_first_search('S', 'G', graph, heuristic)
print('Shortest Path:', path)
```
